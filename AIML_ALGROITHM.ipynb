```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
upload  = files.upload()
```

```python
data = pd.read_csv('cgpa_package_200.csv')
```

```python
data
```

```python
plt.scatter(data['CGPA'],data['Package_LPA'])
plt.xlabel('CGPA')
plt.ylabel('Package_LPA')
```

```python
X = data.iloc[:,0:1]
Y = data.iloc[:,-1]
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size = 0.2,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
X_test
```

```python
Y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(data['CGPA'],data['Package_LPA'])
plt.plot(X_test['CGPA'],lr.predict(X_test),color = 'red')
plt.xlabel('CGPA')
plt.ylabel('Package_LPA')
```

```python
m = lr.coef_
m
```

```python
c = lr.intercept_
c
```
